## 1940. 주몽
- https://www.acmicpc.net/problem/1940

```python
import sys

# input = sys.stdin.readline

N = int(input())
M = int(input())
A = sorted(list(map(int, input().split())))

cnt = 0
l, r = 0, N-1

while l < r:
    if A[l] + A[r] == M:
        cnt += 1
        l += 1
        r -= 1
    elif A[l] + A[r] < M:
        l += 1
        
    elif A[l] + A[r] > M:
        r -= 1
        
print(cnt)
```

# 2003번. 수들의 합 2

- N개의 수로 된 수열 A[1], A[2], …, A[N] 이 있다. 이 수열의 i번째 수부터 j번째 수까지의 합 A[i] + A[i+1] + … + A[j-1] + A[j]가 M이 되는 경우의 수를 구하는 프로그램을 작성하시오.

```python
from itertools import combinations

N, M = map(int, input().split())
a = list(map(int, input().split()))

cnt = 0

for i in range(1, len(a) + 1):
    for j in combinations(a, i):
        print(j)
        if sum(j) == M:
            cnt += 1
print(cnt)
```

```python
# https://www.acmicpc.net/problem/2003

N, M = map(int, input().split())
A = sorted(list(map(int, input().split())))
print(A)
cnt = 0
start = 0

while start != N-1:
for i in range(1, N):
    print(sum(A[start:i+1]))
    if sum(A[start:i+1]) == M:
        cnt += 1
    start += 1
        
print(cnt)
```

```python
# https://www.acmicpc.net/problem/2003

N, M = map(int, input().split())
A = sorted(list(map(int, input().split())))
cnt = 0
start = 0

while start != N-1:
    total = A[start]
    for i in range(1, N):
        total += A[i]
        if total == M:
            cnt += 1
    start += 1

print(cnt)
```

```python
N, M = map(int, input().split())
A = sorted(list(map(int, input().split())))
cnt = 0
end = 0

for start in range(N-1):
    total = A[start]
    while end < N-1:
        total += A[end]
        end += 1
        if total == M:
            cnt += 1
print(cnt)
```
