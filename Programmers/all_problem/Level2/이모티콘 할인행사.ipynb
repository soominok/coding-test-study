```python
# https://wondev.tistory.com/152
# 결과 성공

def solution(users, emoticons):
    answer = [0, 0]
    sale_list = [10, 20, 30, 40]
    discounts = []
    
    def r(arr, depth, l):
        if depth == l:
            discounts.append(arr)
            return
        for i in range(4):
            r(arr + [sale_list[i]], depth+1, l)
            
    r([], 0, len(emoticons))
    
    for discount in discounts:
        emo_plus, emo_tot = 0, 0
        
        for user in users:
            total = 0
            for i in range(len(discount)):
                if user[0] <= discount[i]:
                    total += int(emoticons[i] * ((100-discount[i]))/100)
                    
            if total >= user[1]:
                emo_plus += 1
            else:
                emo_tot += total
                
        if answer[0] < emo_plus:
            answer = [emo_plus, emo_tot]
            
        elif answer[0] == emo_plus:
            if answer[1] <= emo_tot:
                answer = [emo_plus, emo_tot]
                
    return answer
```
