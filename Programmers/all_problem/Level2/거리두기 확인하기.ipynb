```python
# 통과!
def solution(places):
    answer = []
    
    for place in places:
        idx_p = [(y, x) for y in range(5) for x in range(5) if place[y][x] == 'P']
        
        check = 1
        for y1, x1 in idx_p:
            for y2, x2 in idx_p:
                d = abs(y1 - y2) + abs(x1 - x2)
                if d == 0 or d > 2:
                    continue
                if d == 1:
                    check = 0
                elif y1 == y2 and place[y1][int((x1+x2)/2)] != 'X':
                    check = 0
                elif x1 == x2 and place[int((y1+y2)/2)][x1] != 'X':
                    check = 0
                elif y1 != y2 and x1 != x2:
                    if place[y2][x1] != 'X' or place[y1][x2] != 'X':
                        check = 0
                        
        answer.append(check)
    return answer
```
