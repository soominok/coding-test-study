# <문제 1> 시각 - 완전 탐색(Brute Forcing) 문제

### 가능한 경우의 수를 모두 검사해보는 탐색 방법

```python
# H 입력 받기
h = int(input())

count = 0
for i in range(h + 1):
    for j in range(60):
        for k in range(60):
            # 매 시각 안에 '3'이 포함되어 있다면 카운트 증가
            # 시, 분, 초를 하나로 쭉 나열해서 확인
            if '3' in str(i) + str(j) + str(k):
                count += 1
                
print(count)
```

# <문제 2> 왕실의 나이트 - 시뮬레이션 & 완전 탐색

### 2차원 좌표 이용하는 구현 문제 유형

```python
# 현재 나이트의 위치 입력받기
input_data = input()
row = int(input_data[1])
column = int(ord(input_data[0])) - int(ord('a')) + 1

# 나이트가 이동할 수 있는 8가지 방향 정의
steps = [(-2, -1), (-1, -2), (1, -2), (2, -1), (2, 1), (1, 2), (-1, 2), (-2, 1)]

# 8가지 방향에 대하여 각 위치로 이동이 가능한지 확인
result = 0
for step in steps:
    # 이동하고자 하는 위치 확인
    next_row = row + step[0]
    next_column = column + step[1]
    # 해당 위치로 이동이 가능하다면 카운트 증가
    if next_row >= 1 and next_row <= 8 and next_column >= 1 and next_column <= 8:
        result += 1
        
print(result)
```

# <문제 3> 문자열 재정렬 - Facebook 인터뷰 기출

### 수와 숫자의 차이를 알아야 함!
### * 수 : 100, 1000과 같은 정수
### * 숫자 : 0~9의 한글자로 이루어진 글자

```python
data = input()
result = []
value = 0

# 문자를 하나씩 확인하며
for x in data:
    # 알파벳인 경우 결과 리스트에 삽입
    if x.isalpha():
        result.append(x)
    # 숫자는 따로 더하기
    else:
        value += int(x)
        
# 알파벳을 오름차순으로 정렬
result.sort()


# 숫자가 하나라도 존재하는 경우 가장 뒤에 삽입
if value != 0:
    result.append(str(value))
    
#  최종 결과 출력(리스트를 문자열로 변환하여 출력)
print(''.join(result))
```

## <예시 유형 1 - 시뮬레이션 유형> 상하좌우
- 시뮬레이션 유형: 일련이 명령에 따라서 개체를 차례대로 이동시키는 문제

```python
n = int(input())
x, y = 1, 1
plans = input().split()
print(plans)

dx = [0, 0, -1, 1]
dy = [-1, 1, 0, 0]
move_types = ['L', 'R', 'U', 'D']

for plan in plans:
    print(plan)
#     print(x, y)

    for i in range(len(move_types)):
        if plan == move_types[i]:
            nx = x + dx[i]
            ny = y + dy[i]
#             print(nx, ny)
            
    if nx < 1 or ny < 1 or nx > n or ny > n:
        continue
        
    x, y = nx, ny
    
    
print(x, y)
```

## <예시 유형 3> 왕실의 나이트

- 2차원 좌표를 이용하는 "구현" 문제임!

## <예시 유형 4> 문자열 재정렬

```python
data = input()
result = []
value = 0

for x in data:
    if x.isalpha():
        result.append(x)
    else:
        value += int(x)

result.sort()

if value != 0:
    result.append(str(value))

print(''.join(result))
```
